```python
def tic_tac_toe():
    # Initialize the board
    board = [[False for _ in range(3)] for _ in range(3)]
    
    # Players: 'X' and 'O'
    players = ['X', 'O']
    current_player = 0  # 0 for X, 1 for O
    
    print("Welcome to Tic-Tac-Toe!")
    print("You are Player", players[current_player])
    print("Enter a number between 1-9 to mark your move")
    
    while True:
        # Display the board
        display_board(board)
        
        # Get current player's move
        while True:
            try:
                move = int(input("Enter your move (1-9): ")) - 1
                if 0 <= move < 9:
                    break
                else:
                    print("Invalid move. Please enter a number between 1 and 9.")
            except ValueError:
                print("Please enter a valid number.")
        
        # Check if the cell is already taken
        row = move // 3
        col = move % 3
        if board[row][col]:
            print(f"Error: Cell ({row}, {col}) is already taken!")
            continue
        
        # Update the board with the current player's symbol
        board[row][col] = players[current_player]
        
        # Check for a win
        winner = check_winner(board, players[current_player])
        
        if winner:
            print(f"{winner} wins!")
            break
        
        # Check for draw condition
        if len([cell for row in board for cell in row if not cell]) == 0:
            print("It's a draw!")
            break
        
        # Switch players
        current_player = 1 - current_player

def display_board(board):
    for i in range(3):
        print()
        for j in range(3):
            if (j != 0 or i != 0) and not board[i][j]:
                print("   ", end="")
            elif board[i][j]:
                print(f" {board[i][j]} ", end="")
            else:
                print("   ", end="")
        print()

def check_winner(board, symbol):
    # Check rows
    for i in range(3):
        if all(board[i][j] == symbol for j in range(3)):
            return True
    
    # Check columns
    for j in range(3):
        if all(board[i][j] == symbol for i in range(3)):
            return True
    
    # Check diagonals
    if board[0][0] == symbol and board[1][1] == symbol and board[2][2] == symbol:
        return True
    if board[0][2] == symbol and board[1][1] == symbol and board[2][0] == symbol:
        return True
    
    # If none, no winner
    return False

# Start the game
tic_tac_toe()
```

```python
import tkinter as tk

# Define a function to create and configure the main window
def create_main_window():
    # Create the main window
    window = tk.Tk()
    
    # Set the window's title and size
    window.title("Tic-Tac-Toe")
    window.geometry('600x600')  # Adjust the size to better fit the grid
    
    # Return the main window
    return window

# Define a function to create and configure the board frame
def create_board_frame(window):
    # Create the board frame within the main window
    board_frame = tk.Frame(window)
    board_frame.grid(row=1, column=0, sticky='ew')
    
    # Set a default background color for the window, for example white
    windowbg = "white"  # You need to define windowbg or assign a value to it
    
    # Configure the frame's background and padding
    board_frame.config(bg=windowbg, padx=20, pady=20)
    
    # Add grid row and column weights to make the cells expand
    for i in range(3):
        window.grid_rowconfigure(i + 1, weight=1)
        window.grid_columnconfigure(i, weight=1)
    
    return board_frame

# Define a function to create the cells grid
def create_cells(grid, row, col, turn_label):
    # Create a cell widget within the specified grid coordinates
    cell = tk.Button(grid, text="", font=("Arial", 40), width=5, height=2, command=lambda: cell_click(row, col, turn_label))
    cell.grid(row=row, column=col, sticky='ew')
    
    # Configure cell properties here if necessary
    return cell

# Function to check for a win
def check_win():
    global current_player
    # Check rows, columns, and diagonals for a win
    for i in range(3):
        # Check rows
        if cells_grid[i][0]['text'] == cells_grid[i][1]['text'] == cells_grid[i][2]['text'] != "":
            return True
        # Check columns
        if cells_grid[0][i]['text'] == cells_grid[1][i]['text'] == cells_grid[2][i]['text'] != "":
            return True
    
    # Check diagonals
    if cells_grid[0][0]['text'] == cells_grid[1][1]['text'] == cells_grid[2][2]['text'] != "":
        return True
    if cells_grid[0][2]['text'] == cells_grid[1][1]['text'] == cells_grid[2][0]['text'] != "":
        return True
    
    return False

# Function to handle cell click and update the board
def cell_click(row, col, turn_label):
    global current_player  # Access the global current_player variable
    
    # Check if the cell is empty
    if cells_grid[row][col]['text'] == "":
        # Update the clicked cell with the current player's symbol
        cells_grid[row][col].config(text=current_player)
        
        # Check if the current player wins
        if check_win():
            turn_label.config(text=f"Player {current_player} Wins!")
            # Disable all buttons after a win
            for i in range(3):
                for j in range(3):
                    cells_grid[i][j].config(state="disabled")
            return
        
        # Switch turns
        if current_player == "X":
            current_player = "O"
        else:
            current_player = "X"
        
        # Update the turn label
        turn_label.config(text=f"Player {current_player}'s Turn")

# Main application loop
def main():
    global current_player
    current_player = "X"  # Start with player X
    
    # Create and configure the main window
    window = create_main_window()
    
    # Create the board frame
    board_frame = create_board_frame(window)
    
    # Add a label to display the current player's turn
    turn_label = tk.Label(window, text="Player X's Turn", font=("Arial", 20))
    turn_label.grid(row=0, column=0, sticky='ew', pady=20)
    
    # Initialize a 3x3 grid of cells
    global cells_grid
    cells_grid = []
    for i in range(3):
        row_cells = []
        for j in range(3):
            cell = create_cells(board_frame, i, j, turn_label)
            row_cells.append(cell)
        cells_grid.append(row_cells)
    
    # Start the application loop to handle events and update UI
    window.mainloop()

# Run the main application when the module is loaded
if __name__ == "__main__":
    main()
```
